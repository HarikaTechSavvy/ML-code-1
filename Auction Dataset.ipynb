```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder=LabelEncoder()
auction['Lot_Status']=labelencoder.fit_transform(auction['Lot_Status'])
auction
```

```python
#IQR for SalePrice
#we Already know the quartile values from the describe
#IQR=Q3-Q1(inter quartile region)
IQR_SP=1779000-429000
IQR_SP
```

```python
Lower_bound=429000-(1.5*IQR_SP) #lower whisker
Upper_bound=1779000+(1.5*IQR_SP)# upper whisker
print(Lower_bound,Upper_bound)
```

```python
x=auction['lot_EstimateHigh']#x denotes input
y=auction['Sale_Price']#y denotes output
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=1)
```

```python
x_train=x_train.values.reshape(-1,1)
```

```python
from sklearn import linear_model as lm
model=lm.LinearRegression()
results=model.fit(x_train,y_train)
```

```python
model
```

```python
results
```

```python
x_test=x_test.values.reshape(-1,1)
```

```python
predictions=model.predict(x_test)
```

```python
predictions
```

```python
accuracy=model.score(x_train,y_train)
print('Accuracy of the model:',accuracy)
```

```python
print('intercept',model.intercept_)
print('slope:',model.coef_)
```
